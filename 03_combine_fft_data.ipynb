```python
fifteen_min_rain=pd.DataFrame()
for month in range(1,6):
    df=pd.read_parquet(f'./fft/Bayer_hourly_drizzle_month{month}.parquet')
    fifteen_min_rain=pd.concat([fifteen_min_rain, df])
    
    
```

```python
fifteen_min_rain.columns=['time', 'device', 'precip']
```

# combine hourly rain with distrometer data

```python
Bayer_rain_fft=pd.merge(fifteen_min_rain, distometer_fft_rain, on=['time', 'device'], how='inner')
```
